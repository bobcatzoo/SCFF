## Stochastic Calculus for Finance II Continuous-Time Models

### 1.6 Change of Measure

example of 1.6.6 Change of measure for a normal random variable

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.05
x = np.arange(-10, 10, dx)
y = np.exp(- 0.5 * x**2) / np.sqrt(2 ** np.pi)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
ax.set_title('Probability density function of Normal distribution')
ax.grid(axis='both')
ax.plot(x, y, color='b', linewidth=2)
```

```python
cdf = np.cumsum(y * dx)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
ax.grid(axis='both')
ax.plot(x, y, label="pdf of of Normal distribution")
ax.plot(x, cdf, label="cdf of Normal distribution")

ax.set_xlabel("X")
ax.set_ylabel("Probability Values")
ax.set_title("pdf and cdf of Normal distribution")
plt.legend()
```

```python
theta = 1
z = np.exp(theta * x - 0.5 * theta ** 2)
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 12))

ax[0].grid(axis='both')
ax[0].plot(x, z, label="z = np.exp(theta * x - 0.5 * theta ** 2)")
ax[0].set_ylabel("the Values of z")
ax[0].set_title("pdf and cdf of Normal distribution")

ax[1].grid(axis='both')
ax[1].plot(x, np.log(z), label="theta * x - 0.5 * theta ** 2")
ax[1].set_xlabel("X")
ax[1].set_ylabel("the Values of log(z)")


plt.legend()
```

```python
len(z)
```

```python
x_220 = x[0:220]
z_220 = z[0:220]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (6, 10))
ax.grid(axis='both')
ax.plot(x_220, z_220, label="z = np.exp(theta * x - 0.5 * theta ** 2")

ax.set_xlabel("X")
ax.set_ylabel("the Values of z_x")
ax.set_title("pdf and cdf of Normal distribution")
plt.legend()
```

```python
z_x = np.exp(- 0.5 * (x-theta)**2) / np.sqrt(2 ** np.pi)
cdf_z = np.cumsum(z_x * dx)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
ax.grid(axis='both')
ax.plot(x, y, label="pdf of Normal distribution")
ax.plot(x, z_x, label="pdf of z_x")

ax.set_xlabel("X")
ax.set_ylabel("Probability Values")
ax.set_title("pdf and cdf of Normal distribution")
plt.legend()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
ax.grid(axis='both')

ax.plot(x, cdf, label="cdf of Normal distribution")
ax.plot(x, cdf_z, label="cdf of z_x")

ax.set_xlabel("X")
ax.set_ylabel("Probability Values")
ax.set_title("pdf and cdf of Normal distribution")
plt.legend()
```
